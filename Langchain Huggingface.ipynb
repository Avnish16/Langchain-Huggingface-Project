```python
urls = [
    'https://sell.amazon.in/seller-blog/what-is-ecommerce',
    'https://www.bigcommerce.com/articles/ecommerce/',
    'https://en.wikipedia.org/wiki/E-commerce#:~:text=E%2Dcommerce%20(electronic%20commerce),services%20or%20over%20the%20Internet.'
]
```

```python
from langchain_community.document_loaders import UnstructuredURLLoader
loaders = UnstructuredURLLoader(urls=urls)
data = loaders.load()
```

```python
data
```

```python
from langchain.text_splitter import CharacterTextSplitter

text_splitter = CharacterTextSplitter(separator='\n',
                                      chunk_size=4000,
                                      chunk_overlap=400)


docs = text_splitter.split_documents(data)
```

```python
docs
```

```python
len(docs)
```

```python
import pickle
import faiss
from langchain.vectorstores import FAISS
from langchain.embeddings import HuggingFaceEmbeddings

embeddings = HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")
```

```python
embeddings
```

```python
vectorStore_huggingface = FAISS.from_documents(docs, embeddings)

with open("faiss_store_huggingface.pkl", "wb") as f:
    pickle.dump(vectorStore_huggingface, f)
```

```python
with open("faiss_store_huggingface.pkl", "rb") as f:
    VectorStore = pickle.load(f)
```

```python
VectorStore
```

```python
from transformers import pipeline
qa_pipeline = pipeline("question-answering", model="deepset/roberta-base-squad2")
```

```python
from langchain.chains import RetrievalQAWithSourcesChain
from langchain.chains.question_answering import load_qa_chain
from langchain import HuggingFaceHub
```

```python
class HuggingFaceWrapper:
    def __init__(self, pipeline):
        self.pipeline = pipeline

    def __call__(self, inputs, *args, **kwargs):
        response = self.pipeline(question=inputs['question'], context=inputs['context'])
        return response['answer']
```

```python
llm = HuggingFaceWrapper(qa_pipeline)
```

```python
llm
```

```python
from langchain.chains import RetrievalQAWithSourcesChain
```

```python
class CustomRetrievalQAWithSourcesChain:
    def __init__(self, llm, retriever):
        self.llm = llm
        self.retriever = retriever

    def __call__(self, inputs, return_only_outputs=False):
        docs = self.retriever.get_relevant_documents(inputs['question'])
        context = ' '.join([doc.page_content for doc in docs])
        response = self.llm({'question': inputs['question'], 'context': context})
        if return_only_outputs:
            return {'answer': response}
        return {'answer': response, 'sources': docs}
```

```python
chain = CustomRetrievalQAWithSourcesChain(llm=llm, retriever=VectorStore.as_retriever())
```

```python
chain({"question": "What is ecommerce used for?"}, return_only_outputs=True)
```

```python
chain({"question": "What is ecommerce business?"}, return_only_outputs=True)
```
